```python
model = tf.keras.models.load_model("Pass-strength-Predictor")
print("Model loaded.")
```

## **Function to test Password**

```python
def enter_pass():
  password = input("Enter Password :") # GETTING THE PASSWORD
  length = len(password) # FETCHING THE PASSWORD
  chr, digit, symbol = 0, 0, 0
  for i in password:
    if i.isalpha():
      chr += 1
    elif i.isdigit():
      digit += 1
    else:
      symbol += 1

  input_data = [length, chr, digit, symbol] # THE DATA TO GIVE THE MODEL FOR PREDICTING

  strength = model.predict([input_data]) # MAKING PREDICTIONS
  strength = np.argmax(strength) # GETTING THE PREDICTED VALUE FROM A PROBABILITY LIST

  # GIVING THE OUTPUT
  string = 'Your password is'
  if strength == 0:
    print(string, "Weak")
  elif strength == 1:
    print(string, "Medium")
  else:
    print(string, "Strong")
```

```python
enter_pass()
```
